```python
import cv2
from darkflow.net.build import TFNet
```

```python
options = {
    "model": "cfg/yolo.cfg",
    "load": "bin/yolov2.weights",
    "threshold": 0.3
}

tfnet = TFNet(options)
```

```python
video = cv2.VideoCapture("testing.mp4.mp4")

while video.isOpened():
    ret, frame = video.read()

    if not ret:
        break

    results = tfnet.return_predict(frame)

    for result in results:
        tl = (
            result['topleft']['x'],
            result['topleft']['y']
        )

        br = (
            result['bottomright']['x'],
            result['bottomright']['y']
        )

        label = result['label']

        cv2.rectangle(frame, tl, br, (0, 255, 0), 2)

        cv2.putText(
            frame,
            label,
            (tl[0], tl[1] - 10),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.6,
            (255, 255, 255),
            2
        )

    cv2.imshow("YOLOv2 Video Detection", frame)

    if cv2.waitKey(1) & 0xFF == 27:  # ESC key
        break

video.release()
cv2.destroyAllWindows()
```

```python
video = cv2.VideoCapture("testing.mp4.mp4")

width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = video.get(cv2.CAP_PROP_FPS)

out = cv2.VideoWriter(
    "detected_video.avi",
    cv2.VideoWriter_fourcc(*'XVID'),
    fps,
    (width, height)
)

while video.isOpened():
    ret, frame = video.read()

    if not ret:
        break

    results = tfnet.return_predict(frame)

    for result in results:
        tl = (result['topleft']['x'], result['topleft']['y'])
        br = (result['bottomright']['x'], result['bottomright']['y'])

        cv2.rectangle(frame, tl, br, (0, 255, 0), 2)

        cv2.putText(
            frame,
            result['label'],
            (tl[0], tl[1]-10),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.6,
            (255, 255, 255),
            2
        )

    out.write(frame)

video.release()
out.release()
```

```python
import cv2
import matplotlib.pyplot as plt

cap = cv2.VideoCapture("testing.mp4.mp4")

ret, frame = cap.read()

results = tfnet.return_predict(frame)

for r in results:
    tl = (r['topleft']['x'], r['topleft']['y'])
    br = (r['bottomright']['x'], r['bottomright']['y'])

    cv2.rectangle(frame, tl, br, (0,255,0), 2)
    cv2.putText(frame, r['label'], tl,
                cv2.FONT_HERSHEY_SIMPLEX,
                0.7, (255,255,255), 2)

plt.figure(figsize=(12,8))
plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()

cap.release()
```
